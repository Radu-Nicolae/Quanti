```python
LLMs = ['LLama2-7B', 'Granite-7B', 'Mistral-7B']
# two matching colors, one to represent a qunatized version of the LLM one to represent the unquantized version

```

```python
# vertical bar plot with four bars for each LLM
# bars: accuracy of raw LLM, accuracy of quantized LLM, energy consumption of raw LLM, energy consumption of quantized LLM
# scale y-axis to 0-100 for accuracy and 0-10 for energy consumption
# color scheme: blue for accuracy (in 2 shades), orange for energy consumption (in 2 shades)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = {
    'LLM': ['LLama2-7B', 'Granite-7B', 'Mistral-7B'],
    'Accuracy Raw': [85, 80, 78],
    'Accuracy Quantized': [83, 78, 75],
    'Energy Raw': [8, 7, 6],
    'Energy Quantized': [5, 4, 3]
}

df = pd.DataFrame(data)
x = np.arange(len(df['LLM']))  # the label locations
width = 0.2  # the width of the bars
fig, ax1 = plt.subplots(figsize=(10, 6))
bars1 = ax1.bar(x - width, df['Accuracy Raw'], width, label='Accuracy Raw', color='blue')
bars2 = ax1.bar(x, df['Accuracy Quantized'], width, label='Accuracy Quantized', color='lightblue')
ax1.set_ylabel('Accuracy (%)', color='blue')
ax1.set_ylim(0, 100)
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks(x)
ax1.set_xticklabels(df['LLM'])
ax1.legend(loc='upper left')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
bars3 = ax2.bar(x + width, df['Energy Raw'], width, label='Energy Raw', color='orange')
bars4 = ax2.bar(x + 2*width, df['Energy Quantized'], width, label='Energy Quantized', color='lightcoral')
ax2.set_ylabel('Energy Consumption (Joules)', color='orange')
ax2.set_ylim(0, 10)
ax2.tick_params(axis='y', labelcolor='orange')
ax2.legend(loc='upper right')
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('LLM Performance and Energy Consumption')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Larger, IEEE-friendly typography
plt.rcParams.update({
    'font.size': 16,
    'font.family': 'serif',
    'axes.linewidth': 1.2,
    'axes.labelsize': 20,
    'axes.titlesize': 22,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 14,
    'figure.titlesize': 22
})

data = {
    'LLM': ['Llama2-7B', 'Granite-7B', 'Mistral-7B'],
    'Accuracy Raw': [85, 80, 78],
    'Accuracy Quantized': [83, 78, 75],
    'Energy Raw': [8, 7, 6],
    'Energy Quantized': [5, 4, 3]
}

df = pd.DataFrame(data)
x = np.arange(len(df['LLM']))
width = 0.18

fig, ax1 = plt.subplots(figsize=(10, 5))

# Colors
accuracy_raw_color = '#1f4e79'      # Dark blue
accuracy_quant_color = '#5b9bd5'    # Light blue
energy_raw_color = '#d2691e'        # Orange
energy_quant_color = '#ff6b6b'      # Light coral

# Accuracy bars (left axis)
ax1.bar(x - 1.5*width, df['Accuracy Raw'], width, label='Accuracy Raw',
        color=accuracy_raw_color, alpha=0.9)
ax1.bar(x - 0.5*width, df['Accuracy Quantized'], width, label='Accuracy Quantized',
        color=accuracy_quant_color, alpha=0.9)

ax1.set_ylabel('Accuracy [%]')
ax1.set_ylim(0, 100)
ax1.set_xticks(x)
ax1.set_xticklabels(df['LLM'])
ax1.grid(True, alpha=0.3, linestyle='--', axis='y')

# Energy bars (right axis)
ax2 = ax1.twinx()
ax2.bar(x + 0.5*width, df['Energy Raw'], width, label='Energy Raw',
        color=energy_raw_color, alpha=0.9)
ax2.bar(x + 1.5*width, df['Energy Quantized'], width, label='Energy Quantized',
        color=energy_quant_color, alpha=0.9)

ax2.set_ylabel('Energy Usage [kWh]')
ax2.set_ylim(0, 10)

# Single legend (4 rows, 1 column) at the top
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2,
           loc='upper center', bbox_to_anchor=(0.5, 1.18),
           ncol=2, frameon=True, fancybox=False, shadow=False,
           borderpad=0.5, labelspacing=0.4, handlelength=1.2,
           columnspacing=0.8, fontsize=16)
ax1.tick_params(axis='x', labelsize=20)
ax1.tick_params(axis='y', labelsize=20)

# Title and layout (reserve room for legend above)
# plt.title('LLM Performance and Energy Consumption', fontweight='bold', pad=12)
fig.tight_layout(rect=[0, 0, 1, 0.88])

# Clean up spines
ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)

plt.savefig("f1.pdf", bbox_inches='tight')
```
